```python
downriver_pop= pd.read_csv("Demographics of Downriver - Population.csv")
downriver_pop.head()
```

```python
# Map 2 Cities
#River Rouge
#Wyandotte
#Gibraltar
#Monroe 

import folium

map_center = [42.221622, -83.485153] #Centered in Van Buren Charter Township 
downriver_cities = folium.Map(location=map_center, zoom_start=9, tiles='CartoDB positron')


cities = [
    {"name": "River Rouge", "coords": [42.2731, -83.1349]},
    {"name": "Wyandotte", "coords": [42.2142, -83.1499]},
    {"name": "Gibraltar", "coords": [42.0953, -83.1919]},
    {"name": "Monroe", "coords": [41.9164, -83.3977]}
]

for city in cities:
    folium.CircleMarker(
        location=city["coords"],
        radius=8,  
        color='black', 
        fill=True,
        fill_color='black',  
        fill_opacity=1,  
        popup=f"<b>{city['name']}</b>"
    ).add_to(downriver_cities)


downriver_cities
```

```python
downriver_cities.save('downriver_cities_map.html')
```
